```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
plt.plot(x,y)
plt.savefig("images\\fig1.png")
plt.show()
```

```python
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
plt.plot(x,y)
plt.title('sales of samsung phone from 2015 to 2022')
plt.xlabel('year')
plt.ylabel('number of unit sales')
plt.savefig("images\\fig2.png")
plt.show()
```

# Compare Multiple Dataset

```python
plt.rcParams['axes.prop_cycle'].by_key()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.savefig("images\\fig6.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,x,y2,x,y3,'o:r') #mlc(marker linestyle color)
plt.savefig("images\\fig7.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,'o-.b',x,y2,'o--g',x,y3,'o:r')
plt.savefig("images\\fig8.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,x,y2,x,y3,color='teal',linestyle='--',marker='o')
plt.savefig("images\\fig9.png")
plt.show()
```

```python
plt.figure(figsize=(8,8))
x=('22/01/2020','23/01/2020','24/01/2020','25/01/2020','26/01/2020','27/01/2020')
# y1=(555,654,941,1434,2118,2927)
y2=(17,18,26,42,56,82)
y3=(28,30,36,39,52,61)
plt.title('Covid Deaths and Recovery day-wise')
plt.xlabel('Date')
plt.ylabel('Number of people')
plt.annotate('deaths',(3,50),color='red')
plt.annotate('recovered',(2,40),color='green')
# plt.ylim(20,3000)
plt.plot(x,y2,'o-r',x,y3,'o:g')
plt.savefig("images\\fig10.png")
plt.show()
```

# change thickness of line, marker, marker color, alpha

```python
plt.figure(figsize=(8,8))
x=('22/01/2020','23/01/2020','24/01/2020','25/01/2020','26/01/2020','27/01/2020')
y2=(17,18,26,42,56,82)
y3=(28,30,36,39,52,61)
plt.title('Covid Deaths and Recovery day-wise')
plt.xlabel('Date')
plt.ylabel('Number of people')
plt.annotate('deaths',(3,50),color='red')
plt.annotate('recovered',(2,40),color='green')
# plt.ylim(20,3000)
plt.plot(x,y2,'o-r',x,y3,'o:g',lw='3',ms='6',mfc='blue',alpha=0.8)
plt.savefig("images\\fig11.png")
plt.show()
```

# Customization of figure object, x label and y label

```python
plt.figure(figsize=(8,8),facecolor='#989898',edgecolor='black',linewidth='10')
x=('22/01/2020','23/01/2020','24/01/2020','25/01/2020','26/01/2020','27/01/2020')
y2=(17,18,26,42,56,82)
y3=(28,30,36,39,52,61)
plt.title('Covid Deaths and Recoveries',{'color':'red','size':30,'backgroundcolor':'black','verticalalignment': 'center'},pad=30,loc="center")
plt.xlabel('Date',{'color':'red','size':20},labelpad=10)
plt.ylabel('Number of people',{'color':'red','size':20},labelpad=10)
ax=plt.axes()
ax.set_facecolor('#080E14')
plt.annotate('deaths',(3,50),color='red',size=14)
plt.annotate('recovered',(2,39),color='green',size=14)
# plt.ylim(20,3000)
plt.plot(x,y2,'o:r',x,y3,'o-g')
plt.savefig("images\\fig12.png")
plt.show()
```
