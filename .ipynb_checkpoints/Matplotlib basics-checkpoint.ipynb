```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
plt.plot(x,y)
plt.title('sales of samsung phone from 2015 to 2022')
plt.xlabel('year')
plt.ylabel('number of unit sales')
plt.show()
```
