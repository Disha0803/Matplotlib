```python
plt.figure(figsize=(8,8))
x=('22/01/2020','23/01/2020','24/01/2020','25/01/2020','26/01/2020','27/01/2020')
y2=(17,18,26,42,56,82)
y3=(28,30,36,39,52,61)
plt.title('Covid Deaths and Recovery day-wise')
plt.xlabel('Date')
plt.ylabel('Number of people')
plt.annotate('deaths',(3,50),color='red')
plt.annotate('recovered',(2,40),color='green')
# plt.ylim(20,3000)
plt.plot(x,y2,'o-r',x,y3,'o:g',lw='3',ms='6',mfc='blue',alpha=0.8)
plt.savefig("images\\fig11.png")
plt.show()
```

# Customization of figure object, x label and y label

```python
plt.figure(figsize=(8,8),facecolor='#989898',edgecolor='black',linewidth='10')
x=('22/01/2020','23/01/2020','24/01/2020','25/01/2020','26/01/2020','27/01/2020')
y2=(17,18,26,42,56,82)
y3=(28,30,36,39,52,61)
plt.title('Covid Deaths and Recoveries',{'color':'red','size':30,'backgroundcolor':'black','verticalalignment': 'center'},pad=30,loc="center")
plt.xlabel('Date',{'color':'red','size':20},labelpad=10)
plt.ylabel('Number of people',{'color':'red','size':20},labelpad=10)
ax=plt.axes()
ax.set_facecolor('#080E14')
# plt.grid()
plt.annotate('deaths',(3,51),color='red',size=14)
plt.annotate('recovered',(1,36),color='green',size=14)
# plt.ylim(20,3000)
plt.plot(x,y2,'o:r',x,y3,'o-g',lw=3)
plt.savefig("images\\fig12.png")
plt.show()
```

```python
x=np.arange(1,20)
y1=x**2
y2=x**3
line2,=plt.plot(x,y2,marker='o')
line1,=plt.plot(x,y1,marker='o')
plt.legend([line1,line2],['square','cube'])#highly recommended 
plt.savefig("images\\fig24.png")
plt.show()
```
