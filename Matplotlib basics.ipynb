```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
plt.plot(x,y)
plt.savefig("images\\fig1.png")
plt.show()
```

```python
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
plt.plot(x,y)
plt.title('sales of samsung phone from 2015 to 2022')
plt.xlabel('year')
plt.ylabel('number of unit sales')
plt.savefig("images\\fig2.png")
plt.show()
```

```python
x=np.arange(2015,2022)
y=(5000,6000,8000,4000,2000,10000,25000)
# plt.plot(x,y,marker="o",linestyle=''':''',color="red")
# plt.plot(x,y,marker="o",linestyle='''--''',color="red")
plt.plot(x,y,marker="o",linestyle='''-.''',color="red")
plt.title('sales of samsung phone from 2015 to 2022')
plt.xlabel('year')
plt.ylabel('number of unit sales')
plt.savefig("images\\fig3.png")
plt.show()
```

```python
x=np.arange(1,11)
y=x**3
plt.plot(x,y,marker="o",linestyle='''--''',color="green")
plt.savefig("images\\fig4.png")
plt.show()
```

```python
x=np.arange(1,11)
y=x**2
plt.plot(x,y,marker="o",linestyle=''':''',color="blue")
plt.savefig("images\\fig5.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.savefig("images\\fig6.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,x,y2,x,y3,'o:r') #mlc(marker linestyle color)
plt.savefig("images\\fig7.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,'o-.b',x,y2,'o--g',x,y3,'o:r')
plt.savefig("images\\fig8.png")
plt.show()
```

```python
x=np.arange(1,21)
y1=x**2
y2=x**3
y3=x**4
plt.plot(x,y1,x,y2,x,y3,color='teal',linestyle='--',marker='o')
plt.savefig("images\\fig9.png")
plt.show()
```
